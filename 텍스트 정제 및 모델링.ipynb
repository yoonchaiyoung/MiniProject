# 훈련 데이터, 테스트 데이터로 분할

```
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size = 0.2, random_state=42)
```

## 중복 제거

```
train_data.duplicated().sum()  # 중복 데이터가 있는 지 확인
```

```
train_data = train_data.drop_duplicates(subset=['review'])  # 중복 데이터 제거  
train_data.shape  # 중복 데이터 제거되었는 지 확인  # 행 자체를 삭제해줌
```

## 결측값 제거

```
train_data.isnull().sum()  # 결측값 갯수 확인
```

```
# 현재는 없지만 나중에 있을지도 모르니 null 데이터 제거 코드 사용
train_data = train_data.dropna(how='any')
train_data
```

```
print(train_data.isnull().values.any())  # 결측값 존재여부 확인
```

```
train_data.index  # 중복데이터가 제거되어 index가 빈 값이 생김
# index 재설정
```

```
train_data  # 중복, 결측값 제거된 데이터 확인
```

```
for i in train_data.index:
  print(train_data['review'][i])  # 중복, 결측값 제거 됬는지 확인..
```

## 특수기호, 이모티콘, 숫자 제거
- 한글만 뽑아내자.
- 정규식 이용

```
import re
# 한글만 뽑아내서 리스트 안에 넣자.
only_korean_list = []
for i in train_data.index:
  s = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "", str(train_data['review'][i]))
  only_korean_list.append(s)
```

```
import pandas as pd
# train_data 라는 변수에 저장  # 한글만 있는 후기가 저장됨.
train_data = pd.DataFrame({"review" : only_korean_list})
train_data
```

```
for i in train_data.index:
  print(train_data['review'][i])  # 한글을 제외한 특수기호, 이모티콘들 제거되었는 지 확인
```

## 띄어쓰기, 맞춤법 교정
- PyKoSpacing : 한국어 띄어쓰기 교정 패키지
- Py-Hanspell : 네이버 한글 맞춤법 검사기 기준 띄어쓰기&맞춤법 교정 패키지

- 이 중에서 띄어쓰기 교정과 맞춤법 교정이 둘 다 되는 Py-Hanspell 패키지를 사용하자.
