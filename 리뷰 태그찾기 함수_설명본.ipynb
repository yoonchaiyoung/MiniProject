```
from konlpy.tag import Mecab
# 객체 생성
mecab = Mecab()
# 토큰화
for i in data.index:
  temp = []
  temp = mecab.pos(data['review'][i])  # 형태소 단위로 토큰화
  # print(temp[0][1])
  temp_2 = []
  for j in range(len(temp)):
    if temp[j][1] in ['NNG', 'NNP', 'VA', 'MM', 'XSN', 'XR', 'SH']:  # 일반 명사, 고유 명사, 형용사, 관형사, 일반 부사('MAG'), 명사 파생 접미사, 어근, 한자
      temp_2.append(temp[j][0])
  data['review'][i] = temp_2
data
```

## 불용어 제거

```
brand = file_name.split('-')[0]
category = file_name.split('-')[1]
stop_words = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다', '것'] + [brand, category] + ['신발', '컨버스', '좋', '잘', '신발', '같', '사이즈', '생각', '배송', '발', '구매', '적', '템', '신기', '팬츠']  # 새로 불용어 더 추가할 예정..
```

```
for i in data.index:
  no_stop_words = []
  for j in range(len(data['review'][i])):
    if data['review'][i][j] not in stop_words:
      no_stop_words.append(data['review'][i][j])
  data['review'][i] = no_stop_words
data
```

## 최대 문장 길이 정하기
- 원래는 토큰화 전에 최대 문장 길이를 정해서 잘랐어야 했는 데
- 토큰화시키는 품사의 종류를 한정시켰기 때문에
- 문장의 길이가 충분히 줄어들 것이라고 판단했고,
- 리뷰 같은 경우는 전체 내용이 중요하기 때문에 미리 문장을 자르지 않고
- 토큰화를 시킨 다음에 토큰의 갯수를 한정 시키기로 하자.

```
max_token_count = 30  # 최대 토큰의 갯수를 30개로 한정시키자.
```

```
index_list = []
for i in data.index:
  if len(data['review'][i]) <= max_token_count:  # 최대 토큰의 갯수보다 토큰의 갯수가 적은 행 번호 저장
    index_list.append(i)
data = data.loc[index_list]
data  # 최대 토큰의 갯수보다 토큰의 갯수가 적은 행만 뽑아서 data 변수에 저장
```

# 단어 사전

```
# 단어 사전 만들기 위해 모든 토큰 모으기
total_token = []
for i in data.index:
  for j in range(len(data['review'][i])):
    total_token.append(data['review'][i][j])
print(total_token)
```

```
# 단어 빈도수 체크
# total_token.count('좋')
count = 0
token_count_dictionary = dict()
for token in set(total_token):
  count = total_token.count(token)
  token_count_dictionary[token] = count
print(token_count_dictionary)
```

```
# 단어 사전 딕셔너리의 value값 기준 내림차순 정렬
import operator
print(sorted(token_count_dictionary.items(), key=operator.itemgetter(1), reverse=True))
```

```
token_count_dictionary = sorted(token_count_dictionary.items(), key=operator.itemgetter(1), reverse=True)
print(token_count_dictionary)  # 이게 바로 단어 사전
```

```
tag = []
for i in range(10):
  tag.append(token_count_dictionary[i][0])
print(tag)
```

```
# 태그가 list 형태이기 때문에
# 데이터테이블에 들어가지 않음
# 하나의 문장처럼 만들자
tag_str = ""
for i in range(len(tag)):
  tag_str += tag[i] + ' '
print(tag_str)
```
